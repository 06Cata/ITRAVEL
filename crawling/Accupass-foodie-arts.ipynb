```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time

def scroll_down(driver):
    # Get the current page height
    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        # Scroll down to the bottom of the page
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait for the page to load
        time.sleep(2)

        # Calculate the new page height after scrolling
        new_height = driver.execute_script("return document.body.scrollHeight")

        # If the page height remains the same, no more content is being loaded
        if new_height == last_height:
            break

        # Update the last recorded page height
        last_height = new_height

url = "https://www.accupass.com/search?c=arts,food&q=&s=relevance"
options = webdriver.ChromeOptions()
options.add_argument("incognito")
driver = webdriver.Chrome(options=options)
driver.get(url)

# Scroll down until no more content is loaded
scroll_down(driver)

soup = BeautifulSoup(driver.page_source, "html.parser")
req = soup.prettify()
print(req)

driver.quit()
```

```python
event_list = []

event_elements = soup.find_all("p", class_="EventCard-6ce99c89-event-name")
time_elements = soup.find_all("p", class_="EventCard-41b431a9-event-time")
location_spans = soup.find_all('i', class_='EventPlaceType-4d555702-event-location EventPlaceType-a3ceb105-event-offline')

for event_element, time_element in zip(event_elements, time_elements):
    event_name = event_element.get_text(strip=True)
    event_time = time_element.get_text(strip=True)
    
    location_text = None  # Initialize location_text

    for span in location_spans:
        next_sibling = span.next_sibling
        if next_sibling is not None and next_sibling.name == 'span':
            location_text = next_sibling.get_text(strip=True)
            break  # Break the loop after finding the location_text
   
    event_dict = {"Name": event_name, "Location": location_text, "Time": event_time}
    event_list.append(event_dict)

print(event_list)
```

```python
import json

# Define the output JSON file path
output_file = 'foodie_arts_events_0704.json'

# Write the list to the JSON file with proper encoding
with open(output_file, 'w', encoding='utf-8') as f:
    json.dump(event_list, f, ensure_ascii=False)

print("Data exported to", output_file)
```
