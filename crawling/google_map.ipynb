```python
# 印出附近地點的詳細資料
def get_place_details(place_id):
    return gmaps.place(place_id, language='zh-TW')

for place in response['results']:
    place_id = place['place_id']
    details = get_place_details(place_id)
    print("商家名稱:", details['result']['name'])
    print("地址:", details['result']['formatted_address'])
    print("營業時間:", details['result'].get('opening_hours', {}).get('weekday_text'))
    print("評價:", details['result'].get('rating'))
    print("熱門時段:", details['result'].get('popular_times'))
    print("種類:", details['result'].get('types'))
    print("緯度:", details['result']['geometry']['location']['lat'])
    print("經度:", details['result']['geometry']['location']['lng']) 

    price_level = details['result'].get('price_level')
    if price_level is not None:
        print("平均價位:", "$" * price_level)
    else:
        print("價位資訊不可用")
        
        
    reviews = details['result'].get('reviews', [])
    if reviews:
        print("評論:")
        for review in reviews:
            print("- {}：{}".format(review['author_name'], review['text']))
    
    print("--------------")
```

```python
# direction api
import requests

YOUR_API_KEY ='AIzaSyC9rNA2CMy9Ka1_XCT7OtvDbv-KJm7AOBo'

origin = 'Museum of Natural Science, Taiwan'
destination = 'National Taiwan University'

url = f"https://maps.googleapis.com/maps/api/directions/json?origin={origin}&destination={destination}&key={YOUR_API_KEY}"

payload={}
headers = {}

response = requests.request("GET", url, headers=headers, data=payload)

data = response.json()
if data['status'] == 'OK':
    route = data['routes'][0]
    legs = route['legs'][0]

    # 提取起点和终点
    start_address = legs['start_address']
    end_address = legs['end_address']

    # 提取距离
    distance = legs['distance']['text']

    print("起點:", start_address)
    print("終點:", end_address)
    print("距離:", distance)

    steps = route['legs'][0]['steps']
    for step in steps:
        if 'transit_details' in step:
            transit_details = step['transit_details']
            vehicle_type = transit_details['line']['vehicle']['type']
            line = transit_details['line']['short_name']
            departure_stop = transit_details['departure_stop']['name']
            arrival_stop = transit_details['arrival_stop']['name']
            
            print('交通方式:', vehicle_type.encode('unicode-escape').decode('utf-8'))
            print('線路號:', line)
            print('上車站點:', departure_stop.encode('unicode-escape').decode('utf-8'))
            print('下車站點:', arrival_stop.encode('unicode-escape').decode('utf-8'))

    # 提取开车时间和步行时间
    drive_time = legs['duration']['text']
    walk_time = legs['steps'][0]['duration']['text']

    print("開車時間:", drive_time)
    print("步行時間:", walk_time)
else:
    print("無法獲取路線信息")
```

```python
# translate api
def translate_text(target: str, text: str) -> dict:
    """Translates text into the target language.

    Target must be an ISO 639-1 language code.
    See https://g.co/cloud/translate/v2/translate-reference#supported_languages
    """
    from google.cloud import translate_v2 as translate
    import os
    
    os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = r"C:\Users\catal\Desktop\py\my-project-travel-391915-9bf81bc0cbc3.json"

    translate_client = translate.Client()

    if isinstance(text, bytes):
        text = text.decode("utf-8")

    # Text can also be a sequence of strings, in which case this method
    # will return a sequence of results for each text.
    result = translate_client.translate(text, target_language=target)

    print("Text: {}".format(result["input"]))
    print("Translation: {}".format(result["translatedText"]))
    print("Detected source language: {}".format(result["detectedSourceLanguage"]))

    return result

# 示例用法
target_language = "zh-TW"
text = "Hello, how are you?"

translate_text(target_language, text)
```
